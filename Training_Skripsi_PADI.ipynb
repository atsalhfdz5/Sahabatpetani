```python
import os

def cari_results_csv(start_path):
    ditemukan = False
    for root, dirs, files in os.walk(start_path):
        if 'results.csv' in files:
            path_lengkap = os.path.join(root, 'results.csv')
            print(f"➔ Ketemu! Salin path ini:\n{path_lengkap}\n")
            ditemukan = True
    if not ditemukan:
        print(f"Tidak ditemukan file 'results.csv' di dalam folder: {start_path}")

# 1. Cek di Google Drive Anda
print("Mencari di Google Drive...")
cari_results_csv('/content/drive/MyDrive/SKRIPSI_YOLOV11')

# 2. Cek di penyimpanan lokal Colab (jika training dijalankan di sesi ini)
print("Mencari di penyimpanan lokal Colab...")
cari_results_csv('/content/runs')
```
